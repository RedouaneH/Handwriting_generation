```python
from IPython.display import Image as ipython_image, display
    
from generate_handwriting import generate_handwriting

output_path = 'img/hello_world.png'
text = "Hello World!"
bias = 2
color = "red"
style = 9
stroke_width = 2

generate_handwriting(
    text=text,
    stroke_width=stroke_width,
    bias=bias,
    color=color,
    style=style,
    output_path=output_path
)

display(ipython_image(filename=output_path))
```

# STYLES

```python


style_ids = [i for i in range(10)] # there are 211 styles possible from 0 to 210
output_path="img/output.png"

for style_id  in style_ids:
    
    print(f"style_id {style_id}")
    
    generate_handwriting(
        "Daily Meeting",
        output_path="img/output.png",
        style=style_id,
        bias=2,         
        color="black",
        stroke_width=2
    )
    
    display(ipython_image(filename=output_path))
    

```

# WIDTH

```python

widths = [0.5, 1, 1.5, 2, 2.5, 3]
output_path="img/output.png"

for width  in widths:
    
    print(f"width {width}")
    generate_handwriting(
        "Daily Meeting",
        output_path="img/output.png",
        style=1,
        bias=2,         
        color="black",
        stroke_width=width
    )
    display(ipython_image(filename=output_path))
    
```

# Bias

#### High bias $\rightarrow{}$ Low variance, Low bias $\rightarrow{}$ High variance

```python

biases = [0, 0.25, 0.5]
output_path="img/output.png"

for bias  in biases:
    
    print(f"bias = {bias}")
    for i in range(3):
        generate_handwriting(
            "Daily Meeting",
            output_path="img/output.png",
            style=20,
            bias=bias,         
            color="black",
            stroke_width=2
        )
        display(ipython_image(filename="img/output.png"))
    
```
